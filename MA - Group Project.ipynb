```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
ma = pd.read_csv("ma_lga.csv")
ma.saledate=pd.to_datetime(ma.saledate)
ma.set_index('saledate')
ma.head()
```

```python
plt.plot(ma.saledate, ma.MA)
plt.title('Moving average 2007-2016')
plt.xlabel("Saledate")
plt.ylabel("Price")
plt.legend(["Moving average"])
plt.show()
```
